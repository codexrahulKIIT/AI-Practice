```python
# prompt: create a pandas dataframe using dictionary where take 3 colum name age salary and 5 rows values

import pandas as pd

data = {'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Emily'],
        'age': [25, 30, 22, 45, 38],
        'salary': [50000, 60000, 48000, 80000, 75000]}

df = pd.DataFrame(data)
df
```

```python
# @title Name

from matplotlib import pyplot as plt
import seaborn as sns
df.groupby('Name').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python

# Add a new column 'department' to the DataFrame
df['department'] = ['Sales', 'Marketing', 'Engineering', 'HR', 'Finance']
df
```

```python


# Rename the 'age' column to 'experience'
df = df.rename(columns={'age': 'experience'})
df
```

```python

# Drop the 'department' column
df = df.drop('department', axis=1)
df
```

```python
# prompt: you select multiple columns from a dataframe

# Select multiple columns
selected_columns = df[['Name', 'experience']]
selected_columns
```

```python

# Retrieve the row with index 2 (the third row)
row_index = 2  # Example index
retrieved_row = df.iloc[row_index]
retrieved_row
```

```python


# Group the DataFrame by the 'Name' column and calculate the size of each group
grouped_by_name = df.groupby('Name').size()

# Print the result
grouped_by_name
```
